Ahora haremos uso de la funcion merge de los dos dataframes: df1 y df2, con las columnas 'Pais' y 'Aficion principal' como los valores del parametro 'on' 

```python
pd.merge(df1, df2, on=['Pais', 'Aficion principal'])
```

Ademas hacemos uso del parametro 'how' y sus variantes: inner, outer, left y right

```python
pd.merge(df1, df2, on=['Pais', 'Aficion principal'], how='inner')
```

```python
pd.merge(df1, df2, on=['Pais', 'Aficion principal'], how='outer')
```

```python
pd.merge(df1, df2, on=['Pais', 'Aficion principal'], how='left')
```

```python
pd.merge(df1, df2, on=['Pais', 'Aficion principal'], how='right')
```
