```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Establecemos la semilla inicial para la generacion de secuencias de numeros aleatorios 
np.random.seed(125)

columnas = ["Genero", "Edad", "Estatura (cm)", "Profesion", "Ciudad"]

# Generamos valores aleatorios en cada columna
genero = np.random.choice(["M", "F"], size=50)
edad = np.random.randint(20, 65, size=50)
estatura = np.random.randint(148, 192, size=50)
profesion = np.random.choice(["Ingeniero", "Medico", "Abogado", "Economista", "Contador", 
                              "Administrador", "Profesor", "Artista"], size=50)
ciudad = np.random.choice(["Nagoya", "Osaka", "Kyoto", "Yokohama", "Tokio", 
                           "Hokaido", "Okinawa"], size=50)

datos = list(zip(genero, edad, estatura, profesion, ciudad))
df = pd.DataFrame(datos, columns=columnas)
df
```

La principal diferencia entre loc e iloc se da en cómo acceden a los datos en un DataFrame. 
loc utiliza etiquetas, mientras que iloc utiliza índices enteros. La elección entre ellos 
depende de si desea seleccionar datos por etiquetas específicas o por posiciones numéricas.
