```python
import pandas as pd
df = pd.read_csv('./pepsico_products.csv') # Dataframe obtenido de kaggle

df = df.head(50)
df
```

Con el siguiente codigo contamos cuantos productos existen por categoria.

```python
df['Category'].value_counts()
```

Agrupamos de acuerdo a la columna 'Category' y obtenemos el grupo de 'Snack'

```python
df.groupby('Category').get_group('Snack')
```

Agrupamos de acuerdo a la columna 'Category' y obtenemos el grupo de 'Beverages'

```python
df1 = df.groupby('Category').get_group('Beverages')
df1
```

Ahora ordenamos de acuerdo a la columna 'Year Launched', por default de
manera ascendente. 

```python
df1.sort_values(by='Year Launched')
```

Ahora ordenamos de acuerdo a la columna 'Year Launched' de manera descendente.

```python
df1.sort_values(by='Year Launched', ascending=False)
```

Ordenamos por indice.

```python
df1.sort_index()
```

Establecemos como indice a la columna 'Year Launched'

```python
df1.set_index('Year Launched')
```

Ahora volvemos a usar el dataframe inicial

```python
df
```

Filtrado de elementos usando conectores logicos; en este caso filtrar los productos cuya categoria 
sea 'Snack' y cuyo ao de lanzamiento (Year Launched) sea posterior al ao de 1960.

```python
df[(df['Category'] == 'Snack') & (df['Year Launched'] > 1960)]
```

Mediante el uso de regex buscamos los productos ('Product Name') en las cuales contengan 
la letra ('A')

```python
df[df['Product Name'].str.contains('A')]
```

Usando regex hallar todos los productos ('Product Name') cuyo nombre contenga
la letra ('S') o la letra ('M').

```python
df[df['Product Name'].str.contains('^[SM]', case=True, regex=True)]
```

Usando regex hallar todos los productos ('Product Name') cuyo nombre contenga
tanto la letra ('S') como la letra ('M'); osea las 2 a la vez.

```python
df[df['Product Name'].str.contains('(?=.*S)(?=.*M)', case=True, regex=True)]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') contengan la palabra 'na' con
minusculas, para ello usaremos el parametro case=True, case (por case sensitive) al dar el
valor de True solo buscara los productos cuyos nombres contengan 'na' pero con minusculas.

```python
df[df['Product Name'].str.contains('na', case=True)]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') contengan la palabra 'na' pero
ya sea en mayusculas o minusculas, es decir ambos casos, solamente basta con usar el parametro
case y dar el valor de False  

```python
df[df['Product Name'].str.contains('na', case=False)]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') terminan con la letra ('a').

```python
df[df['Product Name'].str.contains('a$', case=True)]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') terminan con la palabra ('da').

```python
df[df['Product Name'].str.contains('da$', case=True)]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') terminan con la letra ('e') o
terminan con la letra ('o').

```python
df[df['Product Name'].str.contains('[eo]$', case=True)]
```

Otra manera de realizar la busqueda anterior es usando el conector logico "o" (|)

```python
df[df['Product Name'].str.contains('e$') | df['Product Name'].str.contains('o$')]
```

Ahora buscaremos los productos cuyo nombre ('Product Name') no terminan con la letra ('e') ni
terminan con la letra ('o'); es decir el complemento de la busqueda anterior.

```python
df[~df['Product Name'].str.contains('[eo]$', case=True)]
```

Ahora vamos a ordenar el dataframe por el ao de lanzamiento (Year Launched)

```python
df.sort_values(by='Year Launched')
```

Vemos que del dataframe anterior dos o mas productos con el mismo ao de lanzamiento, no estan ordenados 
alfabeticamente; para ello en el argumento (by) agregamos la columna 'Product Name'

```python
df.sort_values(by=['Year Launched', 'Product Name'])
```

Ahora vamos a ordenarlo alfabeticamente.

```python
df.sort_values(by='Product Name')
```

Con el metodo set_index cambiamos de indice del original a 'Year Launched'

```python
df.set_index('Year Launched')
```
